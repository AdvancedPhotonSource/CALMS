```python
def recursive_string_search(directory, string, hits):
  for file in os.listdir(directory):
    full_file_path = os.path.join(directory, file)
    if os.path.isfile(full_file_path):
      with open(full_file_path, 'r') as f:
        contents = f.read()
        if string.lower() in contents.lower():
          print(f'Found string "{string}" in file "{full_file_path}"')
          hits.extend(contents.splitlines())
          
    else:
      recursive_string_search(full_file_path, string, hits)
  
```
